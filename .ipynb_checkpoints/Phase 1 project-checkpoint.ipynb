### Creating a copy of the raw data

```python
df_cleaned = df.copy()
```

#### This helps us to avoid later issues incase the copy data is destroyed

### Data cleaning and handling missing data

```python
df_cleaned = df.dropna()                # Removes rows with missing values
df_filled = df.fillna(0)                # Replaces missing values with 0


df_numeric = df.select_dtypes(include=['number'])
df_filled_mean = df_numeric.fillna(df_numeric.mean())  # Replaces missing values with column mean

print("Original", df)
print("\n")
print("Cleaned", df_cleaned)
print("\n")
print("Replaced with 0",df_filled)
print("\n")
print("Replaced with mean",df_filled_mean)
```

### Previewing our cleaned data

```python
df_cleaned.columns
```

### Checking whether there are any duplicates

```python
df_cleaned.duplicated().value_counts()
```

#### Checking for duplicates in rows and removing them is essintial because it can give the wrong insight

```python
df_cleaned["Number.of.Engines"] = pd.to_numeric(
    df_cleaned["Number.of.Engines"],
    errors="coerce"
)
```

#### This is useful when your column may have mixed data types like strings and numbers

### Overview of the injury severity

```python
df_cleaned["Injury.Severity"].value_counts()
```

####  We select the column for analysis, then count how many times each unique value appears. The output shows the severity of the accidents  


### Aircraft category VS Risk

```python
df_cleaned.groupby("Aircraft.Category")["Injury.Severity"].value_counts()
```

####  Groups the dataset by aircraft type, counts the number of occurences of each injury severity with each aircraft category. The output show which aircraft is fatal and is non fatak

### Engines VS Fatalities

```python
df_cleaned.groupby("Number.of.Engines")["Total.Fatal.Injuries"].mean()
```

#### Groups the dataset by the number of engines each aircraft has then calculates the average number of fatal injuries for each engine category. The output shows which type of aircraft tends to have fatalities

### Calculate the average total fatal injuries for each phase of flight

```python
df_cleaned.groupby("Broad.phase.of.flight")["Total.Fatal.Injuries"].mean()
```

#### Groups the data by the phase of flight calculates the average number of fatal injuries. The output shows which flight phase is prone to fatalities


### Creating a flag that will help us compare the data well

```python
df_cleaned["is fatal"] = df_cleaned["Injury.Severity"].str.contains("Fatal", na=False)
```

#### Creates a new column in the DataFrame. 

### Confirming whether the flag exists

```python
df_cleaned.columns
```

### Understand how common fatal vs non-fatal accidents

```python
import matplotlib.pyplot as plt

df_cleaned["is fatal"].value_counts().plot(kind="bar")
plt.title("Distribution of Fatal vs Non-Fatal Accidents")
plt.xlabel("Fatal Accident")
plt.ylabel("Number of Accidents")
plt.show()
```

### Aircraft category vs fatal risk

```python
df_cleaned.groupby("Aircraft.Category")["is fatal"].mean().plot(kind="bar")
plt.title("Fatal Accident Rate by Aircraft Category")
plt.ylabel("Fatal Accident Rate")
plt.xlabel("Aircraft Category")
plt.show()
```

### Number of engines vs fatal risks

```python
df_cleaned.groupby("Number.of.Engines")["is fatal"].mean().plot(kind="bar")
plt.title("Fatal Accident Rate by Number of Engines")
plt.xlabel("Number of Engines")
plt.ylabel("Fatal Accident Rate")
plt.show()
```

### Phase of flight risk

```python
df_cleaned.groupby("Broad.phase.of.flight")["is fatal"].mean().plot(kind="bar", figsize=(10,5))
plt.title("Fatal Accident Rate by Phase of Flight")
plt.ylabel("Fatal Accident Rate")
plt.xlabel("Phase of Flight")
plt.xticks(rotation=45)
plt.show()
```
